# Running analog quantum simulations on myQLM

myQLM comes with a collection of tools that help you to design, run and analyse analog quantum programs on a dedicated QPU - `QutipQPU`. The analog programs, dubbed schedules and encoded by a `Schedule`, contain a Hamiltonian evolved for some fixed time. If it interacts with the environment one can describe the behaviour of the system via jump operators or stochastic noise.

Overall, this folder contains notebooks that combine different tools provided inside `qat.analog`.
